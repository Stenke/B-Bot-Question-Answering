Or create synthetic QA. The idea behind domain-specific QA datasets is that it should provide the language and answer phrasing specific to your desired use-case and improve performance there. (At least this is what studies have shown - link to the one)

With this set-up, our QA dataset is hooked up to the wikipedia articles associated with them. That way, when someone asks a question, our program searches for the top articles, searches for answers, and returns. It circumvents the need for adding context since the context is already here - we just have to pull it up.

This could be re-purposed for our curated datasets on sensitive questions. Instead of Wiki articles, we have various either articles or transcripts (that we scraped) to search through.
