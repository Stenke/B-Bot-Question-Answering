```python
text_list = []
for link in links:
    
    try:
        # The webpage URL
        url = link

        # HTTP Request
        webpage = requests.get(url, headers=headers)

        # Soup Object containing all data
        soup = BeautifulSoup(webpage.content, "lxml")

        script = soup.find_all("div", attrs={"class":'normal-content'})

        body = script[0].find_all('p', class_='Body')

        for text in body:
            text_list.append(text.text)
#             temp_list = []
#             temp_list = text
#             temp_series = pd.Series(temp_list, index=transcripts.columns)
#             transcripts.append(temp_series, ignore_index=True)
    except:
        print('nope')
        continue
```
